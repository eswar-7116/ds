# `loc` and `iloc`

### `loc`:
1. `stop` index is inclusive
2. Columns are accessed with their names
3. Label-based indexing

### `iloc`:
1. `stop` index is exclusive
2. Columns are accessed with their number
3. Integer-position-based indexing

### Difference between `loc` and `iloc`:

| Feature             | `loc`                 | `iloc`                           |
| ------------------- | --------------------- | -------------------------------- |
| Index slicing stop  | Inclusive             | Exclusive                        |
| Columns accessed by | Column names (labels) | Column positions (integer index) |


## 1. Using `loc`:

```python
df.loc[2]  # Returns column at the specified index.
```

```python
df.loc[2, ['Maths', 'CS']]  # Returns the specified rows from the column at the specified index.
```

```python
df.loc[2][:3]  # Stop index is inclusive as I said earlier
```

```python
df.loc[1:2, ['Maths']]  # Stop roll no. is also inclusive
```

```python
df.loc[1:6:2, ['Maths']]
```

## Using `iloc`:

```python
df.iloc[1:3, [1, 5]]
```

```python
df.iloc[1:3, 1:4]
```

# Manipulating DataFrames

Add/Delete/Modify the data

```python
df["Total Marks"] = 0  # Create a column, "Total Marks" with default value 0 for every row.
df
```

```python
df['Total Marks'] = sum([df['English'], df['Maths'], df['Science'], df['Hindi'], df['CS']])
df
```

```python
df['Total Marks'] = df[['English', 'Maths', 'Science', 'Hindi', 'CS']].sum(axis=1)  # axis = 0 for row and 1 for column
df
```

```python
df.drop(columns="Maths")  # Returns dataframe with no Maths column.
# To remove in-place, set inplace=True. E.g.: df.drop(columns="Maths", inplace=True)
```

```python
df.drop(columns=["Maths", "CS"])
```

# Using loc with conditions

```python
df.loc[df["Maths"] > 80]  # Returns rows with Maths marks > 80
```

```python
df.loc[(df["Maths"] < 60) | (df["CS"] > 80)]  # Returns rows with Maths marks < 60 or CS marks > 80
```

```python
df.loc[(df["Maths"] < 80) & (df["CS"] > 80)]  # Returns rows with Maths marks < 80 and CS marks > 80
```
